```python
import re

def clean_minimal_with_hashtags(text):
    if not isinstance(text, str):
        return ""
    
    # Keep @mentions and #hashtags
    
    # Remove non-ASCII characters (emojis, special symbols)
    text = re.sub(r'[^\x00-\x7F]+', '', text)
    
    # Keep @ and #, remove unwanted symbols
    text = re.sub(r'[^\w\s@#.,!?:/\-]', '', text)
    
    # Normalize whitespace
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text

# Apply version that keeps hashtags
data["Cleaned Post"] = data["Post Text"].apply(clean_minimal_with_hashtags)
```

```python
def build_prompt(text):
    return f"""
You are an AI assistant extracting structured information from customer complaints about Blinkit.

If the complaint is not primarily about Blinkit, return this exact JSON:
{{
  "Issue": null,
  "Product": null,
  "Location": null,
  "Order_ID": null,
  "Sentiment": null,
  "Urgency": null,
  "Resolution_Requested": null
}}

Otherwise, return a valid flat JSON object with the following fields:

- "Issue": Mention all key issues stated in the complaint
- "Product": List all products explicitly mentioned, or null
- "Location": City or region if specified, or null
- "Order_ID": Any order reference if stated, or null
- "Sentiment": One of ["Positive", "Negative", "Neutral"]
- "Urgency": One of ["High", "Medium", "Low"]
- "Resolution_Requested": List all the actions the user is requesting or expecting to resolve the issue

Guidelines:
- Use only the information present in the complaint
- If multiple items apply to any field, include all
- Do not guess or make assumptions
- Keep values short and clean
- Only return valid JSON with no explanation or extra text

Complaint:
\"\"\"{text}\"\"\"

Output:
"""
```

```python
def extract_info(text):
    prompt = build_prompt(text)
    try:
        response = requests.post(
            "http://localhost:11434/api/generate",
            json={
                "model": "llama3",
                "prompt": prompt,
                "stream": False
            }
        )
        return response.json()["response"]
    except Exception as e:
        return str(e)
```

```python
def parse_response(resp):
    try:
        return json.loads(resp)
    except json.JSONDecodeError:
        return {}
```

```python
def is_primary_blinkit_complaint(text: str) -> bool:
    """
    Returns True if Blinkit appears to be the main focus of the complaint.
    """
    if not isinstance(text, str):
        return False

    text_lower = text.lower()
    blinkit_index = text_lower.find("blinkit")
    
    if blinkit_index == -1:
        return False
    
    if blinkit_index < 150:  
        return True

    if text_lower.count("blinkit") >= 2:
        return True

    return False
```

```python
test_data = data.iloc[100:120].copy()

test_data = test_data[test_data["Cleaned Post"].apply(is_primary_blinkit_complaint)].copy()

test_data["llm_raw"] = test_data["Cleaned Post"].apply(extract_info)

test_data["llm_parsed"] = test_data["llm_raw"].apply(
    lambda x: parse_response(x) if x else {}
)

df_extracted = pd.json_normalize(test_data["llm_parsed"]).reset_index(drop=True)
test_data = test_data.reset_index(drop=True)

final_test_df = pd.concat([test_data, df_extracted], axis=1)

final_test_df.head()
```

```python
import pandas as pd
import json
import time

full_data = data.copy()

print(len(full_data))
full_data = full_data[full_data["Cleaned Post"].apply(is_primary_blinkit_complaint)].copy()
print(len(full_data))
full_data = full_data.reset_index(drop=True)

def extract_info_safe(text):
    try:
        result = extract_info(text)
        time.sleep(1)
        return result
    except Exception as e:
        print(f"❌ Error in extract_info: {e}")
        return None

def parse_response_safe(resp):
    try:
        return parse_response(resp)
    except Exception as e:
        print(f"❌ JSON parse error: {e}")
        return {}

full_data["llm_raw"] = None
full_data["llm_parsed"] = None

for idx in range(len(full_data)):
    text = full_data.loc[idx, "Cleaned Post"]
    
    raw = extract_info_safe(text)
    parsed = parse_response_safe(raw) if raw else {}

    full_data.at[idx, "llm_raw"] = raw
    full_data.at[idx, "llm_parsed"] = parsed

    product = parsed.get("Product", None)
    print(f"✅ Row {idx} | Product: {product}")

df_extracted = pd.json_normalize(full_data["llm_parsed"]).reset_index(drop=True)
final_df = pd.concat([full_data.reset_index(drop=True), df_extracted], axis=1)

final_df.head()
```

```python
import pandas as pd
import json
import requests
import re
import time

df = pd.read_csv("overall_opt_new1.csv")

def build_standardization_prompt(issue, product, resolution):
    return f"""
You are an AI assistant responsible for standardizing extracted fields from Blinkit customer complaints.

Here is the extracted data from a complaint:
- Issues: {issue}
- Products: {product}
- Resolution Requested: {resolution}

Your job is to clean and standardize each field based only on the information given. Follow these rules:

Cleaning Rules:
- Correct any typos, inconsistent casing, or plural forms.
- Remove only true duplicates or overlapping items.
- Use clean, short, consistent phrasing.
- For products, use generic product types for all products — not brand names.
- For resolutions, include all valid request the user expresses.

Important:
- DO NOT drop items unless they are clearly invalid or duplicate.
- DO NOT add missing information or guess intent.
- DO NOT include explanations or commentary — return only clean structured data.

Output Format:
Return only a valid JSON object with the exact structure below:
{{
  "Issue": [cleaned issues as short strings],
  "Product": [cleaned products as short strings or null if empty],
  "Resolution_Requested": [cleaned resolutions as short strings]
}}

Ensure the JSON is properly formatted and parsable.
"""

def extract_json(text):
    try:
        match = re.search(r"\{[\s\S]*\}", text)
        return json.loads(match.group(0)) if match else {}
    except Exception as e:
        print("❌ JSON extraction failed:", e)
        return {"Issue": None, "Product": None, "Resolution_Requested": None}

def standardize_row_with_llm(row, idx):
    try:
        prompt = build_standardization_prompt(row["Issue"], row["Product"], row["Resolution_Requested"])
        response = requests.post(
            "http://localhost:11434/api/generate",
            json={"model": "llama3", "prompt": prompt, "stream": False}
        )
        response_text = response.json().get("response", "")
        parsed = extract_json(response_text)

        print(f"✅ Row {idx} | Product: {row['Product']} → Cleaned: {parsed.get('Product', None)}")
        time.sleep(1)  # Optional delay to avoid overload
        return parsed

    except Exception as e:
        print(f"❌ Row {idx} | Request failed: {e}")
        return {"Issue": None, "Product": None, "Resolution_Requested": None}

standardized_outputs = []
for idx in range(len(df)):
    row = df.iloc[idx]
    result = standardize_row_with_llm(row, idx)
    standardized_outputs.append(result)

df_standardized = pd.json_normalize(standardized_outputs)
df = pd.concat([df.reset_index(drop=True), df_standardized.add_suffix("_cleaned")], axis=1)

df[["Product", "Product_cleaned"]].sample(10)
```
