```python
blinkit_keywords = [

    # üì¶ Service / Delivery Issues
    "Blinkit complaint",
    "Blinkit delivery issue",
    "Blinkit not delivered",
    "Blinkit cancelled order",
    "Blinkit wrong item",

    # üí≥ Refund / Payment Issues
    "Blinkit refund issue",
    "Blinkit payment failed",
    "Blinkit charged twice",
    "Blinkit not refunding",

    # ü•¶ Product Quality Issues
    "Blinkit expired product",
    "Blinkit rotten vegetables",
    "Blinkit poor packaging",
    "Blinkit not fresh",

    # üëé Negative Sentiment / Experience
    "Blinkit worst experience",
    "Blinkit terrible service",
    "Blinkit never ordering again",
    "Blinkit horrible service",
    "Blinkit scam",

    # üßë‚Äçüíª Customer Support Issues
    "Blinkit no response",
    "Blinkit support issue",
    "Blinkit help",
    "Blinkit customer care"
]
```

```python
import time
import random
from datetime import datetime, timedelta
from bs4 import BeautifulSoup
from tqdm import tqdm

all_tweets = set()
tweet_list = []

def human_scroll(driver, scroll_count=5):
    for _ in range(scroll_count):
        driver.execute_script("window.scrollBy(0, window.innerHeight / 2);")
        time.sleep(random.uniform(2, 4))

for idx, query in enumerate(tqdm(blinkit_keywords)):
    wr_query = ("%20").join(query.split())

    end_date = datetime.today().strftime('%Y-%m-%d')
    start_date = (datetime.today() - timedelta(days=45)).strftime('%Y-%m-%d')

    link = f"https://x.com/search?q={wr_query}%20until%3A{end_date}%20since%3A{start_date}&src=typed_query&f=top"
    driver.get(link)

    time.sleep(random.uniform(2, 3))

    last_height = driver.execute_script("return document.body.scrollHeight")

    while True:
        human_scroll(driver, scroll_count=random.randint(4, 6))
        time.sleep(random.uniform(2.2, 3.8))

        if random.random() < 0.08:
            time.sleep(random.uniform(10, 20))

        soup = BeautifulSoup(driver.page_source, 'html.parser')
        posts = soup.find_all('div', attrs={'data-testid': 'cellInnerDiv'})

        for post in posts:
            try:
                article = post.find('article', attrs={"data-testid": "tweet"})
                if not article:
                    continue

                p1 = article.find('div', class_="css-175oi2r r-1iusvr4 r-16y2uox r-1777fci r-kzbkwu")
                if not p1:
                    continue

                p2 = p1.find_all('div', recursive=False)
                if len(p2) < 2:
                    continue

                try:
                    complainant_name = p2[0].find_all('a')[0].text.strip()
                except:
                    complainant_name = ""

                try:
                    complainant_id = p2[0].find_all('a')[1].text.strip()
                except:
                    complainant_id = ""

                try:
                    date_posted = p2[0].find_all('a')[2].find('time').get('datetime')
                except:
                    date_posted = ""

                have_img_vid = "Yes" if any(i.find('img') for i in p2) else "No"

                is_reply = "Replying to @" in p2[1].text
                try:
                    written_post = p2[2].text.strip() if is_reply else p2[1].text.strip()
                except:
                    written_post = ""

                stats_index = 4 if is_reply and have_img_vid == "Yes" else \
                              3 if is_reply else \
                              3 if have_img_vid == "Yes" else 2

                try:
                    stats = p2[stats_index].find('div', class_="css-175oi2r r-1kbdv8c r-18u37iz r-1wtj0ep r-1ye8kvj r-1s2bzr4")\
                                           .find_all('div', class_="css-175oi2r r-18u37iz r-1h0z5md r-13awgt0")
                    comments = stats[0].text.strip() if len(stats) > 0 else "0"
                    repost = stats[1].text.strip() if len(stats) > 1 else "0"
                    likes = stats[2].text.strip() if len(stats) > 2 else "0"
                    views = stats[3].text.strip() if len(stats) > 3 else "0"
                except:
                    comments = repost = likes = views = "0"

                try:
                    post_link = "https://x.com" + p2[0].find_all('a')[2].get('href')
                except:
                    post_link = ""

                if post_link and post_link not in all_tweets:
                    all_tweets.add(post_link)
                    tweet_list.append([
                        complainant_name, query, complainant_id, date_posted,
                        written_post, post_link, have_img_vid, comments, repost, likes, views
                    ])
                    print(complainant_id)

            except Exception as e:
                print("Tweet skipped due to structure mismatch:", e)

        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height

    wait_time = random.uniform(30, 60)
    print(f"Sleeping for {wait_time:.2f} seconds before next query...")
    time.sleep(wait_time)
```
